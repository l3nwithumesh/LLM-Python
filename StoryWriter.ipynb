```python
#l3nwithumesh
#issuing a bunch of prompts that build on each other to create a good final outcome
#create a new article from a story skeleton

from openai import OpenAI
from pathlib import Path

client = OpenAI(base_url="http://127.0.0.1:1234/v1", api_key="lm-studio")


messages = [
    {"role": "system", "content": "You are a new article writer."},
]

def askTheLLM(prompt : str):
    messages.append({"role": "user", "content": prompt})
    response = client.chat.completions.create(
        model="google/gemma-3-4b",
        messages=messages
    )
    
    answer = response.choices[0].message.content
    messages.append({"role": "assistant", "content": answer})
    return answer

# Example serial prompts
# first read the file 
from pathlib import Path
resume_string = Path("C:\\Users\\ugaje\\OneDrive\\Documents\\SampleStorySkeleton.txt").read_text()
#print(resume_string)
prompt = "Here is the story skeleton. Identify key characters, story chronology and arrange in chronological order:" + resume_string;
print(askTheLLM(prompt))
print(askTheLLM("Rewrite the story from a foreginer's perspective as it happened to him."))
print(askTheLLM("Optimize it as a small news item no longer than 500 words."))
print(askTheLLM("Now produce a final polished version for a local US news paper."))
```
