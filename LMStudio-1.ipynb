```python
# Define the prompt and parameters
import requests
payload = {
    "model": "google/gemma-3-4b",  # Replace with your desired model name - you can find it in "Model's API Identifier in Dev Console of lMStudio on right bottom
    "prompt": "State the benefits of local LLM system and then tell me a joke", # whatever you want to prompt the LLM with
    "max_tokens": 500, # number of tokens to use
    "temperature": 0.7 # how much intense you want the LLM to be when responding
}

# Send the request
response = requests.post("http://127.0.0.1:1234/v1/completions", json=payload) # get the server connection end point from "the local server is reachable at this address" in LM studio

if response.status_code == 200:
    data = response.json()
    print("Completion Response:")
    print(data.get("choices", [{}])[0].get("text", "No response"))
else:
    print("Error: {response.status_code} - {response.text}")
```
